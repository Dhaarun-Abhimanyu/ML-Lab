```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Sample Data
matches_played = np.array([10, 12, 14, 16, 18])
runs_scored = np.array([287, 300, 297, 350, 345])

# Perform Linear Regression
slope, intercept, _, _, _ = linregress(matches_played, runs_scored)

# Generate trend line data
x_values = np.linspace(min(matches_played), max(matches_played), 100)
print(x_values)
y_values = slope * x_values + intercept

# Create figure
plt.figure(figsize=(8, 5))

# Scatter plot
plt.scatter(matches_played, runs_scored, color='blue', marker='*', label='Data Points')

# Trend line
plt.plot(x_values, y_values, color='blue', linewidth=2, label='Trend Line')

# Labels and Title
plt.xlabel("Number of Matches Played")
plt.ylabel("Number of Runs Scored")
plt.title("IPL Performance: Matches Played vs. Runs Scored (Trend Line)")

# Grid and Legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
data_1 = np.random.normal(loc=100, scale=10, size=200)
data_2 = np.random.normal(loc=90, scale=20, size=200)
data_3 = np.random.normal(loc=80, scale=30, size=200)
data_4 = np.random.normal(loc=70, scale=40, size=200)

df = pd.DataFrame({'data_1': data_1, 'data_2': data_2, 'data_3': data_3, 'data_4': data_4})

box_colors = ['#0000FF', '#00FF00', '#FFFF00', '#FF00FF']

fig, ax = plt.subplots(figsize=(10, 7))

box = ax.boxplot(df.values, vert=False, notch=True, patch_artist=True, 
                 boxprops=dict(color='black', linewidth=1.5),  
                 whiskerprops=dict(color='#8B008B', linestyle=':', linewidth=1.5),  
                 capprops=dict(color='#8B008B', linewidth=2),  
                 medianprops=dict(color='red', linewidth=3),  
                 flierprops=dict(marker='D', color='#e7298a', alpha=0.5))  

for patch, color in zip(box['boxes'], box_colors):
    patch.set_facecolor(color)

ax.set_yticklabels(['data_1', 'data_2', 'data_3', 'data_4'])

ax.set_title('Customized box plot', fontsize=14)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

states = ['California', 'Texas', 'New York', 'Florida', 'Illinois']
num_students = [120, 95, 75, 85, 65]

plt.figure(figsize=(5, 5))

plt.pie(num_students, labels=states, autopct='%.2f%%')

plt.title("Students Enrolled from Different States")

plt.show()
```

```python
import matplotlib.pyplot as plt

time_intervals = [(20, 25), (25, 30), (30, 35), (35, 40), (40, 45), (45, 50), (50, 55)]
num_machines = [3, 5, 11, 8, 5, 2, 1]
colors = ['#FF9999', '#FFCC99', '#FFFF99', '#CCFF99', '#99FFFF', '#CC99FF', '#FFC0CB']

fig, ax = plt.subplots(figsize=(8, 5))

for (start, end), count, color in zip(time_intervals, num_machines, colors):
    ax.bar(start, count, width=end-start, color=color, edgecolor='black')

ax.set_xlabel("Time in Minutes")
ax.set_ylabel("Number of Machines")
ax.set_title("Histogram: Machines Working vs. Time Intervals")

ax.set_xticks([t for interval in time_intervals for t in interval])

ax.yaxis.grid(True, linestyle='--', color='gray', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

vehicles = ['Bicycle', 'Car', 'Bus', 'Train', 'Airplane']
average_speeds = [15, 80, 60, 100, 600]
colors = ['blue', 'green', 'red', 'purple', 'orange']
plt.figure(figsize=(8, 5))

plt.bar(vehicles, average_speeds, color=colors, edgecolor='black')

plt.xlabel("Vehicle Type")
plt.ylabel("Average Speed (km/h)")
plt.title("Average Speed of Different Vehicles")
plt.minorticks_on()
plt.grid(which='major', axis='x', linestyle='dotted', alpha=0.7)

plt.grid(which='minor', axis='y', linestyle='dotted', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_A = [45, 52, 58, 60, 62, 71]
product_B = [30, 35, 45, 47, 50, 55]
product_C = [50, 48, 46, 45, 44, 42]

plt.figure(figsize=(12, 6))

plt.plot(months, product_A, marker='o', linestyle='-', color='blue', label='Product A')
plt.plot(months, product_B, marker='s', linestyle='-', color='red', label='Product B')
plt.plot(months, product_C, marker='^', linestyle='-', color='green', label='Product C')

plt.xlabel("Months")
plt.ylabel("Sales (in thousands)")
plt.title("Product Sales Comparison - First Half 2024")

plt.ylim(0, max(product_A + product_B + product_C)+10)
plt.xlim(1.69, len(months) )

plt.legend()

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
